## Q.10. What is the maximum duration of a Movie/Show on Netflix ?

```python
netflix[:3]
```

```python
netflix.Duration.unique()
```

In above output we can see that the Duration is mentioned in min and also in Seasons so i will actually create new column with min entries only

```python
netflix.Duration[netflix.Duration.str.contains('Season')]
```

```python
# stripping thr string part, converting it into integer as it will be easy to find maximum value from it

netflix.Duration[~netflix.Duration.str.contains('Season')].str.strip(' min').astype(int).max()
```
